Analyze the ‘math score’ field.

```python
import seaborn as sns
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 10))
fig.suptitle('Analyzing the Maths Score distribution.', fontsize=20)
# Math score PDF
sns.kdeplot(x='math score', data=data, ax=ax1)
ax1.title.set_text('PDF')
# Math score CDF
sns.ecdfplot(x='math score', data=data, ax=ax2)
ax2.title.set_text('CDF')
# Math score QQ Plot
qq_plot = stats.probplot(x=data['math score'], plot=plt)
ax3.title.set_text('QQ Plot')
```

Analyze the Reading Score

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 10))
fig.suptitle('Analyzing the Reading Score distribution.', fontsize=20)
# Reading score PDF
sns.kdeplot(x='reading score', data=data, ax=ax1)
ax1.title.set_text('PDF')
# Reading score CDF
sns.ecdfplot(x='reading score', data=data, ax=ax2)
ax2.title.set_text('CDF')
# Reading score QQ Plot
qq_plot = stats.probplot(x=data['reading score'], plot=plt)
ax3.title.set_text('QQ Plot')
```

analyze the “writing score”

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 10))
fig.suptitle('Analyzing the Writing Score distribution.', fontsize=20)
# Reading score PDF
sns.kdeplot(x='writing score', data=data, ax=ax1)
ax1.title.set_text('PDF')
# Reading score CDF
sns.ecdfplot(x='writing score', data=data, ax=ax2)
ax2.title.set_text('CDF')
# Reading score QQ Plot
qq_plot = stats.probplot(x=data['writing score'], plot=plt)
ax3.title.set_text('QQ Plot')
```

The scores from a Gender

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 10))
fig.suptitle('Comparing Male & Female scores for all three subjects:', fontsize=20)
# Math score
sns.boxplot(x='gender', y='math score', data=data, ax=ax1)
ax1.title.set_text('Math score vs Gender')
# Reading Score
sns.boxplot(x='gender', y='reading score', data=data, ax=ax2)
ax2.title.set_text('Reading score vs Gender')
# Writing Score
sns.boxplot(x='gender', y='writing score', data=data, ax=ax3)
ax3.title.set_text('Writing score vs Gender')
```

Observation regarding the Scores vs Gender plots:

In Math, on average, Male students have scored more than Female students.

one student that scored zero in Maths is a Female.

In Reading, female students have done better than male students. Also, the student that scored the lowest marks is a female.

In Writing, females have performed better than males and the student that scored the lowest in this subject is also a female.
