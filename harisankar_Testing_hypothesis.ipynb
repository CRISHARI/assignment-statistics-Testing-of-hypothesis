```python
import numpy as np
from scipy import stats
data1=data['Sales_After_digital_add(in $)']
mean=231123
t_val,p_val=stats.ttest_1samp(data1,mean)
alpha=0.05
if p_val<alpha:
   print('Reject null hypothesis')
else:
    print('do not reject null hypothesis')
```
